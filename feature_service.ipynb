# From ArcGIS Feature Service to Pandas DataFrame

```python
import pandas as pd
import requests
```

```python
url = "https://maps-cartes.ec.gc.ca/arcgis/rest/services/CESI_FGP_All_Layers/MapServer/6/query"
params = {
    "where":"1=1",
    "outfields": "*",
    "geometry": True,
    "f": "json"
}
response = requests.get(url, params)
data = response.json()
```

```python
features = data['features']
df_full = pd.json_normalize(features)
df_full.head()
```
